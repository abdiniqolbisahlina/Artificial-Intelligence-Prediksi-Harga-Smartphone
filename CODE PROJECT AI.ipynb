**ENCODED**

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Kolom-kolom yang ingin di-encode
columns_to_encode = ['brand', 'os', 'battery_type', 'video_720p', 'video_1080p', 'video_4K', 'video_8K',
                     'video_30fps', 'video_60fps', 'video_120fps', 'video_240fps', 'video_480fps', 'video_960fps', 'resolution']

# Menggunakan LabelEncoder untuk setiap kolom
label_encoder = LabelEncoder()
for col in columns_to_encode:
    phone[col + '_encoded'] = label_encoder.fit_transform(phone[col])

# Menghapus kolom-kolom asli yang sudah di-encode
phone = phone.drop(columns=columns_to_encode)

# Drop kolom 'phone_name'
phone = phone.drop(['phone_name'], axis=1)

# Menampilkan beberapa baris pertama dari DataFrame yang sudah di-encode dan dimodifikasi
print(phone.head())

# Simpan DataFrame yang sudah di-encode ke dalam file CSV
phone.to_csv('encodedbiasa.csv', index=False)

print("DataFrame yang sudah di-encode berhasil disimpan ke dalam file CSV.")
```

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

# Baca data CSV
data = pd.read_csv('normalizeddatabaru55.csv')

# Pisahkan fitur dan target
columns_to_drop = [ 'video_720p_encoded','video_1080p_encoded','video_4K_encoded','video_8K_encoded', 'video_30fps_encoded', 'video_60fps_encoded',
       'video_120fps_encoded', 'video_240fps_encoded', 'video_480fps_encoded','video_960fps_encoded']

X = data.drop(columns=columns_to_drop)
Y = data['price(USD)']

# Inisialisasi objek seleksi fitur
selector = SelectKBest(score_func=f_classif, k=9) #implementasi ANOVA

# Lakukan seleksi fitur
X_new = selector.fit_transform(X, Y) #implementasi ANOVA

# Dapatkan indeks fitur yang terpilih
selected_features_indices = selector.get_support()

# Dapatkan nama fitur yang terpilih
selected_features = X.columns[selected_features_indices]

# Tampilkan hasil seleksi fitur
print("Fitur Terpilih:", selected_features)

# Simpan fitur terpilih ke dalam DataFrame baru
selected_features_df = X[selected_features]

# Gabungkan DataFrame fitur terpilih dengan kolom target
selected_data = pd.concat([selected_features_df, Y], axis=1)

# Simpan DataFrame dengan fitur terpilih dan kolom target ke dalam file CSV
selected_data.to_csv('selected_features55.csv', index=False)

print("Data dengan fitur terpilih berhasil disimpan ke dalam file CSV.")
```

**SPLIT DATA**

```python
# Membaca data
data = pd.read_csv('selected_features55.csv')
print(data.head())
```

```python
# Memisahkan fitur (X) dan variabel target (y)
X = data.drop('price(USD)', axis=1)
y = data['price(USD)']
```

```python
from sklearn.model_selection import train_test_split
# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Menampilkan bentuk (shape) dari matriks data latih dan data uji, serta vektor target
print("Bentuk X_train:", X_train.shape)
print("Bentuk X_test:", X_test.shape)
print("Bentuk y_train:", y_train.shape)
print("Bentuk y_test:", y_test.shape)
```

**MULTIPLE LINIER REGRESSION**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
```

```python
# Scaling data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Membuat model Multiple Linear Regression
mlr_model = LinearRegression()

# Melatih model pada data yang telah diproses
mlr_model.fit(X_train_scaled, y_train)

# Melakukan prediksi pada data uji
y_pred = mlr_model.predict(X_test_scaled)
```

```python
# Evaluasi model
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100

# Menampilkan hasil evaluasi
print(f'R-squared (R2): {r2:.17f}')
print(f'Mean Squared Error (MSE): {mse:.17f}')
print(f'Mean Absolute Error (MAE): {mae:.17f}')
print(f'Mean Absolute Percentage Error (MAPE): {mape:.17f}')
```

**RIDGE REGRESSION**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
```

```python
# Penanganan outlier dengan metode IQR (Interquartile Range)
Q1 = X_train.quantile(0.20)
Q3 = X_train.quantile(0.80)
IQR = Q3 - Q1
X_train_no_outlier = X_train[~((X_train < (Q1 - 1.5 * IQR)) | (X_train > (Q3 + 1.5 * IQR))).any(axis=1)]
y_train_no_outlier = y_train[X_train_no_outlier.index]
```

```python
# Scaling data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_no_outlier)
X_test_scaled = scaler.transform(X_test)
```

```python
# Membuat model Ridge Regression dengan pemilihan fitur
ridge_model = Ridge(alpha=0.01)
selector = SelectFromModel(ridge_model)
selector.fit(X_train_scaled, y_train_no_outlier)
X_train_selected = selector.transform(X_train_scaled)
X_test_selected = selector.transform(X_test_scaled)
```

```python
# Melatih model pada data yang telah diproses
ridge_model.fit(X_train_selected, y_train_no_outlier)

# Melakukan prediksi pada data uji
y_pred = ridge_model.predict(X_test_selected)

# Evaluasi model
r2 = ridge_model.score(X_test_selected, y_test)  # Menggunakan metode score untuk mendapatkan R-squared (R2)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100  # Mengalikan dengan 100 untuk mendapatkan nilai dalam persen
```

```python
# Menampilkan hasil evaluasi
print(f'R-squared (R2): {r2:.17f}')  # Menampilkan R-squared (R2) dalam bentuk desimal
print(f'Mean Squared Error (MSE): {mse:.17f}')  # Menampilkan MSE dalam bentuk desimal
print(f'Mean Absolute Error (MAE): {mae:.17f}')  # Menampilkan MAE dalam bentuk desimal
print(f'Mean Absolute Percentage Error (MAPE): {mape:.17f}%')  # Menampilkan MAPE dalam bentuk desimal
```

**LASSO REGRESSION**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectFromModel
```

```python
# Penanganan Outlier dengan metode IQR
Q1 = X_train.quantile(0.20)
Q3 = X_train.quantile(0.80)
IQR = Q3 - Q1
X_train_no_outlier = X_train[~((X_train < (Q1 - 1.5 * IQR)) | (X_train > (Q3 + 1.5 * IQR))).any(axis=1)]
y_train_no_outlier = y_train[X_train_no_outlier.index]
```

```python
# Scaling data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_no_outlier)
X_test_scaled = scaler.transform(X_test)
```

```python
# Membuat model Lasso Regression tanpa pemilihan fitur
lasso_model = Lasso(alpha=0.01)  # Sesuaikan alpha sesuai kebutuhan Anda

# Melatih model pada data yang telah diproses
lasso_model.fit(X_train_scaled, y_train_no_outlier)

# Melakukan prediksi pada data uji
y_pred = lasso_model.predict(X_test_scaled)

# Evaluasi model
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
```

```python
# Menampilkan hasil evaluasi
print(f'R-squared (R2): {r2}')
print(f'Mean Squared Error (MSE): {mse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Absolute Percentage Error (MAPE): {mape}%')
```
